```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/heart.csv')

male_data = df[df['Sex'] == "M"]
female_data = df[df['Sex'] == "F"]

male_disease = male_data['HeartDisease'].value_counts()
female_disease = female_data['HeartDisease'].value_counts()

plt.subplot(1, 2, 1)
plt.pie(male_disease, autopct='%1.1f%%', colors=['orangered','lightgreen'])
plt.title('Heart Disease in Males')

plt.subplot(1, 2, 2)
plt.pie(female_disease, autopct='%1.1f%%', colors=['lightgreen','orangered'])
plt.title('Heart Disease in Females')

plt.show()
```

```python
ata_data = df[df['ChestPainType'] == 'ATA']
ta_data = df[df['ChestPainType'] == 'TA']
nap_data = df[df['ChestPainType'] == 'NAP']
asy_data = df[df['ChestPainType'] == 'ASY']

ATA_disease = ata_data['HeartDisease'].value_counts()
TA_disease = ta_data['HeartDisease'].value_counts()
NAP_disease = nap_data['HeartDisease'].value_counts()
ASY_disease = asy_data['HeartDisease'].value_counts()

plt.subplot(1, 4, 1)
plt.pie(ATA_disease, autopct='%1.1f%%', colors=['lightgreen', 'orangered'])
plt.title('ATA')

plt.subplot(1, 4, 2)
plt.pie(TA_disease, autopct='%1.1f%%', colors=['lightgreen', 'orangered'])
plt.title('TA')

plt.subplot(1, 4, 3)
plt.pie(NAP_disease, autopct='%1.1f%%', colors=['lightgreen', 'orangered'])
plt.title('NAP')

plt.subplot(1, 4, 4)
plt.pie(ASY_disease, autopct='%1.1f%%', colors=['orangered', 'lightgreen'])
plt.title('ASY')

plt.show()
```
